```python
import os
from glob import glob
from tqdm.notebook import tqdm
from dateutil import parser

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import pandas as pd
import geopandas as gpd
```

```python
df_aoi = gpd.read_file('../data/aoi/japan_ver83.shp')
df_aoi = df_aoi[df_aoi["JCODE"].isna()==False]
df_aoi["JCODE"] = df_aoi["JCODE"].astype(int)
df_aoi.plot()
plt.show()
```

```python
df = pd.read_csv("../data/night_light.csv",index_col=0)
df.index = [parser.parse(l) for l in df.index]
```

```python
# 新宿区
jcode = df_aoi[df_aoi["SIKUCHOSON"]=="新宿区"]["JCODE"].astype(str)
df.loc[:,jcode].plot(legend=None)
plt.ylim([0,120])
plt.ylabel("Light emissions [$\mathrm{nWatts·cm^{−2}·sr^{-1}}$]")
plt.show()
```

```python
# 札幌
jcode = df_aoi[(df_aoi["KEN"]=="北海道") & (df_aoi["SIKUCHOSON"]=="中央区")]["JCODE"].astype(str)
df.loc[:,jcode].plot(legend=None)
plt.ylim([0,120])
plt.ylabel("Light emissions [$\mathrm{nWatts·cm^{−2}·sr^{-1}}$]")
plt.show()
```
